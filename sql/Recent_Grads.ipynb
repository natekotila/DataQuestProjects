```python
import sqlite3
conn = sqlite3.connect('jobs.db')
cursor = conn.cursor()
```

```python
query = """
SELECT *
FROM recent_grads;"""

cursor.execute(query)

results = cursor.fetchall()

print(results[:3])
```

```python
query = '''
SELECT Major
FROM recent_grads
;'''

cursor.execute(query)

majors = cursor.fetchall()

print(majors[:3])
```

```python
conn = sqlite3.connect('jobs2.db')

query = '''
SELECT Major
FROM recent_grads
ORDER BY Major DESC
;'''

reverse_alphabetical = conn.execute(query).fetchall()

print(reverse_alphabetical)
```
